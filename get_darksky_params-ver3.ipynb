#### Get weather information using DarkSky API

```python
import requests
import pandas as pd
from collections import namedtuple
from datetime import datetime
```

```python
#Using UNIX time (01-07-2012 00:00:00, 02-07-2012 00:00:00, hourly)

def weather_info(start_t, end_t, lat, lon):
    df_0 = []
    params = [
        "time", #The UNIX time, seconds since midnight GMT on 1 Jan 1970
#         "summary", #Basically sunny, windy, rainy, cloudy,etc-Do we need this?
        "temperature", #Farenheit degrees(later convert to Celsius)
        "precipIntensity", #Volume mm if there is rain or snow
        "precipProbability", #Rain or snow probability
        "windSpeed", # Numerical value representing the wind speed in kilometers per hour
        "windGust", #Wind Gusts is a quick and sudden increase in the speed of wind
#         "windBearing", #Wind direction
#         "cloudCover",#Cloud percentage
#         "visibility",
        "humidity"
    ]
    
    #API_KEY this is jedha's api key. Need to ask.
    #Using lat,lon in Ausgrid data
    #-33.891348,151.191243
    
    HourlySummary = namedtuple("HourlySummary", params)
    
    #Daily
    for time_point in range(start_t,end_t,86400):#24h*3600seconds->1 day
        darksky_url="https://api.darksky.net/forecast/15074402c474e3dab67f7377e1f95519/" + str(lat) + "," + str(lon) + "," + str(time_point) + "?exclude=currently,flags,alerts,minutely,daily"
        response = requests.get(darksky_url)
        data = response.json()
        
        #Hourly
        for i in range(0,24): # 24 hours
            df_0.append(
                HourlySummary(
                    time=data['hourly']['data'][i]['time'],
#                   summary=data['hourly']['data'][i]['summary'],
                    temperature=data['hourly']['data'][i]['temperature'],
                    precipIntensity = data['hourly']['data'][i]['precipIntensity'],
                    precipProbability = data['hourly']['data'][i]['precipProbability'],
                    windSpeed = data['hourly']['data'][i]['windSpeed'],
                    windGust = data['hourly']['data'][i]['windGust'],
#                     windBearing = data['hourly']['data'][i]['windBearing'],
#                   cloudCover = data['hourly']['data'][i]['cloudCover'],
#                   visibility = data['hourly']['data'][i]['visibility'],
                    humidity = data['hourly']['data'][i]['humidity']
                )
            )

    #Convert to DataFrame
    df = pd.DataFrame(df_0, columns=params)

    #Convert UNIX datetime human-readable datetime
    df['time']=df['time'].apply(lambda x: datetime.utcfromtimestamp(int(x)).strftime('%Y-%m-%d %H:%M:%S'))

    #Convert sFarenheit to Celsius
    df['temperature'] = df['temperature'].apply(lambda x: (x-32)*5/9)

    #Write to csv files
    df.to_csv(str(lat) + str(lon) + str(unix_time) + str(duration_h) + ".csv", index=False)
```
