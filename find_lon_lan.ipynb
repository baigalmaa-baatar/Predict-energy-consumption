## Find lattitude and longtitude from postcode in Ausgrid dataset

Find lat, lon on dataset using public file Australian_Post_Codes_Lat_Lon.csv.\
To get historical information from DarkSky, we need lat, lon.

```python
import pandas as pd
```

```python
df_whole=pd.read_csv(r"C:\Users\baiga\Documents\0 Final project\dataset\Australian_Post_Codes_Lat_Lon.csv")
df_postcode=pd.read_csv(r"C:\Users\baiga\Documents\0 Final project\dataset\postcodes.csv", names=['postcode'])
```

```python
df_whole.tail()
```

```python
df_postcode.head()
```

```python
df = df_whole.merge(df_postcode, how="right", on="postcode")
```

```python
df
```

```python
df['postcode'].value_counts()
```

Dropping duplicated rows. In big cities there are many postcodes, but weather is _almost_ same.\
We choose last "lat", "lon" values.

```python
df.drop_duplicates(subset='postcode', keep="last", inplace=True)
df
```

```python
df.drop(columns=["suburb", "state", "dc", "type"], inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
df.to_csv("dataset\lat_lon.csv",index=False)
```
